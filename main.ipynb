```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import os
from glob import glob
%matplotlib inline
```

```python
DATA_DIR = './data'
RUNS_DIR = './runs'
SAVE_DIR = './model'
```

## Load Training Losses

```python
losses = pickle.load(file=open(os.path.join(SAVE_DIR, 'losses.p'),mode='rb'))
```

```python
plt.plot(losses['train'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Training Avg. Loss')
```

```python
plt.plot(losses['test'])
plt.title('Testing Loss')
plt.xlabel('Epoch')
plt.ylabel('Testing Avg. Loss')
```
